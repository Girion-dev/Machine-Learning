```python
def sigmoid(z):
    g = 1 / (1 + np.exp(-z))
    return g

def propagate(w, b, X, Y):
    #TODO change
    m = X.shape[0]
    
    #Forward propagation
    #-------------------------------------------------------------------------------------
    #Get the value of the sample with the current settings
    A = sigmoid(np.dot(w.T, X.T) + b)
    
    #print("Sigmoid: " + str(A))
    #print("Actual label: " + str(Y))
    #print("Calculated label (sigmoid): " + str(A))
    cost = -(1 / m) * np.sum(np.dot(Y, np.log(A).T) + np.dot((1 - Y), np.log(1 - A).T))
    print("Cost: " + str(cost))
    #-------------------------------------------------------------------------------------
    
    #Backward propagation
    #-------------------------------------------------------------------------------------
    dw = (1 / m) * np.dot(X.T, (A - Y).T)
    db = (1 / m) * np.sum(A - Y)
    
    assert(dw.shape == w.shape)
    assert(db.dtype == float)
    cost = np.squeeze(cost)
    assert(cost.shape == ())
    
    grads = {"dw": dw,
             "db": db}
    
    return grads, cost
    
    
    return 0, 0

def train(w, b, X, Y, num_iterations, learning_rate):
    costs = []
    for i in range(num_iterations):
        grads, cost = propagate(w, b, X, Y)
        
        #Get the derivatives from backprop
        dw = grads["dw"]
        db = grads["db"]
        
        #Update values
        w = w - learning_rate * dw
        b = b - learning_rate * db
        
        if i % 10 == 0:
            costs.append(cost)
    
    params = {"w": w,
              "b": b}
    
    grads = {"dw": dw,
             "db": db}
    
    return params, grads, costs
        
    return 0, grads, cost

def model(X_train, Y_train, X_test, Y_test, num_iterations, learning_rate):
    #Initialize parameters
    w = np.zeros((X_train.shape[1], 1))
    b = 0
    
    #Calculate everything after one run
    parameters, grads, costs = train(w, b, X_train, Y_train, num_iterations, learning_rate)
    
    w = parameters["w"]
    b = parameters["b"]
    
    return w, b
```
